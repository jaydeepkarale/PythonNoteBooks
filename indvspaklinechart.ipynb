```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("IndiaVsPak.csv")
```

```python
over = df['Over']
India = df['India']
Pakistan = df['Pakistan']
```

```python
plt.figure(figsize=(8, 8))
plt.title('India vs Pak World T20 Worm Graph', fontsize=20)
plt.xticks(over)
plt.xlabel('Overs', fontsize=14)
plt.ylabel('Runs', fontsize=14)
plt.plot(over, India,'b', label="India")
plt.plot(over, Pakistan,'g', label="Pakistan")
plt.legend()
plt.show()
```
