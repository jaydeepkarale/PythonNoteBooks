```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('DenmarkvsFrance.csv')
```

```python
width=0.25
fig, ax = plt.subplots()
parameters=df['Parameter']
ax.bar(parameters, df['Denmark'], width, label='Denmark',color='r', edgecolor='black')
ax.bar(parameters, df['France'], width, bottom=df['Denmark'],label='France', color='b',edgecolor='black')
ax.tick_params('x',labelrotation=90)
ax.set_title('Denmark Vs France By Numbers')
ax.legend()
plt.show()
```
