```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('RishiSunakApprovalRating.csv')
newdf = df.transpose()
newdf.reset_index(inplace=True)
newdf.columns=newdf.iloc[0]
newdf.drop(index=0,inplace=True)
newdf.rename(columns={'Do you think Rishi Sunak is doing a good job or a bad job as Chancellor of the Exchequer?': 'Month'}, inplace=True)
newdf['Month'] = pd.to_datetime(newdf['Month'])
newdf['Month'] = newdf['Month'].dt.strftime('%Y-%b')
```

```python
good_job = newdf['Good job']
bad_job = newdf['Bad job']
not_sure = newdf['Not sure']
month = newdf['Month']
```

```python
plt.figure(figsize=(8, 8))
plt.title('Rishi Sunak Approval Ratings Over Time', fontsize=20)
plt.xticks(np.arange(0, len(month)+1, 4),rotation = 45)
plt.xlabel('Time Period', fontsize=14)
plt.ylabel('% Ratings', fontsize=14)
plt.plot(month, good_job,'g', label="Good Job", marker='o')
plt.plot(month, bad_job,'r', label="Bad Job",marker='o')
plt.legend()
plt.show()
```
